```python
import numpy as np
import matplotlib.pyplot as plt
import json

# 加载数据
d_json_path = '/content/drive/MyDrive/minist-data-rc-work/minist_dict.json'
with open(d_json_path, 'r') as json_file:
    loaded_data = json.load(json_file)

# 转换为 NumPy 数组
for key, value in loaded_data.items():
    loaded_data[key] = np.array(value)

# 添加噪声的函数
def add_noise(data, noise_level=0.5):
    noise = np.random.normal(0, noise_level, data.shape)  # 添加噪声
    return data + noise

# One-hot 编码函数
def one_hot(y, c):
    y_hot = np.zeros((len(y), c))
    y_hot[np.arange(len(y)), y] = 1
    return y_hot

# Softmax 函数
def softmax(z):
    exp = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp / np.sum(exp, axis=1, keepdims=True)

# 训练模型的函数
def fit(X, y, lr, c, epochs):
    m, n = X.shape
    w = np.random.randn(n, c) * 0.01  # 权重初始化
    b = np.zeros(c)  # 偏置初始化

    losses = []
    accuracy_list = []

    for epoch in range(epochs):
        # 计算预测值
        z = X @ w + b
        y_hat = softmax(z)

        # 计算损失
        y_hot = one_hot(y, c)
        loss = -np.mean(np.log(y_hat[np.arange(len(y)), y]))
        losses.append(loss)

        # 计算梯度
        w_grad = (1/m) * np.dot(X.T, (y_hat - y_hot))
        b_grad = (1/m) * np.sum(y_hat - y_hot, axis=0)

        # 更新权重和偏置
        w -= lr * w_grad
        b -= lr * b_grad

        # 计算准确率
        accuracy = np.mean(np.argmax(y_hat, axis=1) == y)
        accuracy_list.append(accuracy)

        # 每10个epoch输出一次
        if epoch % 10 == 0:
            print(f'Epoch {epoch} => Loss = {loss:.4f}, Accuracy = {accuracy:.4f}')

    return w, b, losses, accuracy_list

# 扩展数据集并加噪声
X_train_extended = []
y_train_extended = []

for _ in range(100):
    for nl in range(10):
        str_digit = f"digit_{nl}"
        inp_signal = loaded_data[str_digit]
        noisy_signal = add_noise(inp_signal, noise_level=0.5)
        X_train_extended.append(noisy_signal.flatten())
        y_train_extended.append(nl)

X_train_extended = np.array(X_train_extended)
y_train_extended = np.array(y_train_extended)

# 打印数据集形状
print(f"X_train_extended shape: {X_train_extended.shape}")

# 训练模型
w, b, losses, accuracy_list = fit(X_train_extended, y_train_extended, lr=0.001, c=10, epochs=100)

# 绘制损失曲线
plt.figure(figsize=(12, 6))
plt.plot(losses)
plt.title('Loss curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()

# 绘制准确率曲线
plt.figure(figsize=(12, 6))
plt.plot(accuracy_list)
plt.title('Accuracy curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()

# 使用训练好的权重进行预测
def predict(X, w, b):
    z = X @ w + b
    y_hat = softmax(z)
    return np.argmax(y_hat, axis=1)

# 预测
y_pred = predict(X_train_extended, w, b)

# 打印预测结果
print(f"Predictions: {y_pred}")

# 计算混淆矩阵
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_train_extended, y_pred)

# 打印混淆矩阵
print("Confusion Matrix with Counts:")
print(conf_matrix)

# 可视化混淆矩阵
plt.figure(figsize=(6, 6), dpi=300)
plt.imshow(conf_matrix, cmap='viridis', interpolation='nearest', aspect='equal')
plt.title("Confusion Matrix for Digits", fontsize=16)
plt.colorbar(label="Counts")
tick_marks = np.arange(10)

plt.xticks(tick_marks, tick_marks, fontsize=12)
plt.yticks(tick_marks, tick_marks, fontsize=12)
plt.xlabel('Predicted Label', fontsize=14)
plt.ylabel('True Label', fontsize=14)

# 显示每个方块的数字
for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{conf_matrix[i, j]}', ha='center', va='center', color='white', fontsize=14)

# 自动调整布局
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import json

# 加载数据
d_json_path = '/content/drive/MyDrive/minist-data-rc-work/minist_dict.json'
with open(d_json_path, 'r') as json_file:
    loaded_data = json.load(json_file)

# 转换为 NumPy 数组
for key, value in loaded_data.items():
    loaded_data[key] = np.array(value)

# 添加噪声的函数
def add_noise(data, noise_level=0.5):
    noise = np.random.normal(0, noise_level, data.shape)  # 添加噪声
    return data + noise

# One-hot 编码函数
def one_hot(y, c):
    y_hot = np.zeros((len(y), c))
    y_hot[np.arange(len(y)), y] = 1
    return y_hot

# Softmax 函数
def softmax(z):
    exp = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp / np.sum(exp, axis=1, keepdims=True)

# ReLU 激活函数
def relu(z):
    return np.maximum(0, z)

# ReLU 的梯度
def relu_grad(z):
    return (z > 0).astype(float)

# Adam优化器的参数更新
def adam_optimizer(w, b, dw, db, m_w, v_w, m_b, v_b, lr, t, beta1=0.9, beta2=0.999, epsilon=1e-8):
    m_w = beta1 * m_w + (1 - beta1) * dw
    v_w = beta2 * v_w + (1 - beta2) * dw**2
    m_b = beta1 * m_b + (1 - beta1) * db
    v_b = beta2 * v_b + (1 - beta2) * db**2

    m_w_hat = m_w / (1 - beta1**t)
    v_w_hat = v_w / (1 - beta2**t)
    m_b_hat = m_b / (1 - beta1**t)
    v_b_hat = v_b / (1 - beta2**t)

    w -= lr * m_w_hat / (np.sqrt(v_w_hat) + epsilon)
    b -= lr * m_b_hat / (np.sqrt(v_b_hat) + epsilon)

    return w, b, m_w, v_w, m_b, v_b

# 训练模型的函数
def fit(X, y, lr, c, epochs):
    m, n = X.shape
    w = np.random.randn(n, c) * 0.01  # 权重初始化
    b = np.zeros(c)  # 偏置初始化

    m_w, v_w = np.zeros_like(w), np.zeros_like(w)
    m_b, v_b = np.zeros_like(b), np.zeros_like(b)

    losses = []
    accuracy_list = []

    for epoch in range(epochs):
        # 计算预测值
        z = X @ w + b
        y_hat = softmax(z)

        # 计算损失
        y_hot = one_hot(y, c)
        loss = -np.mean(np.log(y_hat[np.arange(len(y)), y]))
        losses.append(loss)

        # 计算梯度
        w_grad = (1/m) * np.dot(X.T, (y_hat - y_hot))
        b_grad = (1/m) * np.sum(y_hat - y_hot, axis=0)

        # 使用 Adam 优化器更新权重和偏置
        t = epoch + 1
        w, b, m_w, v_w, m_b, v_b = adam_optimizer(w, b, w_grad, b_grad, m_w, v_w, m_b, v_b, lr, t)

        # 计算准确率
        accuracy = np.mean(np.argmax(y_hat, axis=1) == y)
        accuracy_list.append(accuracy)

        # 每1个epoch输出一次
        if epoch % 1 == 0:
            print(f'Epoch {epoch} => Loss = {loss:.4f}, Accuracy = {accuracy:.4f}')

    return w, b, losses, accuracy_list

# 扩展数据集并加噪声
X_train_extended = []
y_train_extended = []

for _ in range(100):
    for nl in range(10):
        str_digit = f"digit_{nl}"
        inp_signal = loaded_data[str_digit]
        noisy_signal = add_noise(inp_signal, noise_level=0.5)
        X_train_extended.append(noisy_signal.flatten())
        y_train_extended.append(nl)

X_train_extended = np.array(X_train_extended)
y_train_extended = np.array(y_train_extended)

# 打印数据集形状
print(f"X_train_extended shape: {X_train_extended.shape}")

# 训练模型
w, b, losses, accuracy_list = fit(X_train_extended, y_train_extended, lr=0.0001, c=10, epochs=50)

# 绘制损失曲线
plt.figure(figsize=(12, 6))
plt.plot(losses)
plt.title('Loss curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()

# 绘制准确率曲线
plt.figure(figsize=(12, 6))
plt.plot(accuracy_list)
plt.title('Accuracy curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()

# 使用训练好的权重进行预测
def predict(X, w, b):
    z = X @ w + b
    y_hat = softmax(z)
    return np.argmax(y_hat, axis=1)

# 预测
y_pred = predict(X_train_extended, w, b)

# 打印预测结果
print(f"Predictions: {y_pred}")

# 计算混淆矩阵
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_train_extended, y_pred)

# 打印混淆矩阵
print("Confusion Matrix with Counts:")
print(conf_matrix)

# 可视化混淆矩阵
plt.figure(figsize=(6, 6), dpi=300)
plt.imshow(conf_matrix, cmap='viridis', interpolation='nearest', aspect='equal')
plt.title("Confusion Matrix for Digits", fontsize=16)
plt.colorbar(label="Counts")
tick_marks = np.arange(10)

plt.xticks(tick_marks, tick_marks, fontsize=12)
plt.yticks(tick_marks, tick_marks, fontsize=12)
plt.xlabel('Predicted Label', fontsize=14)
plt.ylabel('True Label', fontsize=14)

# 显示每个方块的数字
for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{conf_matrix[i, j]}', ha='center', va='center', color='white', fontsize=14)

# 自动调整布局
plt.tight_layout()
plt.show()
```
